```python
df['target']=wine.target

df.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
hm = sns.heatmap(corr_df, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=df.columns, 
                 xticklabels=df.columns, 
                 cmap="Spectral_r")
plt.show()
```
